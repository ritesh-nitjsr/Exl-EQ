```python
def add_floor_index(x):
    if(x['floor']=="Concourse"):
        x['Floor_index'] = 0
    elif(x['floor']=="Floor 1"):
        x['Floor_index'] = 1
    elif(x['floor']=="Floor 2"):
        x['Floor_index'] = 2
    elif(x['floor']=="Floor 3"):
        x['Floor_index'] = 3
    else:
        x['Floor_index'] = 4
    return x
    
    
    
    

shopper_location = shopper_location.apply(lambda x : add_floor_index(x), axis=1)
```

```python
shopper_location.tail()
```

```python
import math 
from time import time
def add_store_name(x):
    store_name = ""
    min_dist = 10000000000000
    for index,row in store_mapping.iterrows():
        if(row['Floor_Index'] == x['Floor_index']):
            dist = math.sqrt( (x['lat'] - row['latitude'])**2 + (x['lng'] - row['longitude'])**2 )
            if(dist < min_dist):
                store_name = row['Store_Name']
                min_dist = dist
    x['Store_Name'] = store_name
    return x

t0 = time()
shopper_location = shopper_location.apply(lambda x : add_store_name(x), axis=1)
t1 = time()
print t1-t0
```
